```python
import matplotlib.pyplot as plt

# Count of each gender
gender_counts = data['Sex'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'pink'], startangle=140)
plt.title('Gender Distribution in Titanic Dataset')
plt.axis('equal')  # Equal aspect ratio to make the pie circular
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count of each gender, including any category like 'Other'
gender_counts = data['Sex'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    gender_counts, 
    labels=gender_counts.index, 
    autopct='%1.1f%%', 
    startangle=140,
    colors=plt.cm.Set3.colors  # colorful and supports more categories
)
plt.title('Gender Distribution (Including Other, if any)')
plt.axis('equal')  # Make the pie chart round
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count of each gender (including any third gender like 'Other')
gender_counts = data['Sex'].value_counts()

# Custom function to show count and percentage
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count})'
    return my_autopct

# Pie chart with count + percentage
plt.figure(figsize=(7,7))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct=make_autopct(gender_counts),
    startangle=140,
    colors=plt.cm.Pastel2.colors
)
plt.title('Gender Distribution with Count and Percentage')
plt.axis('equal')
plt.show()
```
