```python
# . Univariate Analysis
for col in ['gender' , 'SeniorCitizen' , 'Partner' , 'Dependents' , 'Churn' , 'PaymentMethod']:
    (data[col].value_counts())
    sns.countplot(data = data , x = col)
    plt.show()
```

```python
# . Numerical Distributions
num_cols = ['tenure' , 'MonthlyCharges' , 'TotalCharges' , 'PaymentMethod']
for col in num_cols:
    plt.figure(figsize = (8 , 6))
    sns.histplot(data[col] , kde = True , bins = 30)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Select only numeric columns for correlation heatmap
num_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Compute correlation
corr_matrix = data[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
# Ensure TotalCharges is numeric
if 'TotalCharges' in data.columns:
    data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')

# Encode categorical columns for correlation
data_encoded = data.copy()
```

```python
for col in data_encoded.select_dtypes(include='object').columns:
    if set(data_encoded[col].unique()) <= {'Yes','No'}:
        data_encoded[col] = data_encoded[col].map({'Yes':1,'No':0})

# Drop ID-like columns
if 'customerID' in data_encoded.columns:
    data_encoded = data_encoded.drop(columns=['customerID'])

# One-hot encode remaining categorical columns
data_encoded = pd.get_dummies(data_encoded, drop_first=True)

# Compute correlation matrix
corr_matrix = data_encoded.corr()

# Plot full heatmap
plt.figure(figsize=(18,12))
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.2)
plt.title("Full Feature Correlation Heatmap (Encoded Data)")
plt.show()
```

```python
# Count of each gender
gender_counts = data['gender'].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'pink'], startangle=140)
plt.title('Gender Distribution ')
plt.axis('equal')  # Equal aspect ratio to make the pie circular
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count of each gender (including any third gender like 'Other')
gender_counts = data['gender'].value_counts()

# Custom function to show count and percentage
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        count = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({count})'
    return my_autopct

# Pie chart with count + percentage
plt.figure(figsize=(7,7))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct=make_autopct(gender_counts),
    startangle=140,
    colors=plt.cm.Pastel2.colors
)
plt.title('Gender Distribution with Count and Percentage')
plt.axis('equal')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12, 6)

sns.barplot(
    x="tenure", 
    y="MonthlyCharges", 
    data=data, 
    palette="viridis"
)

plt.title("Average Monthly Charges by Tenure", fontsize=16)
plt.xlabel("Tenure", fontsize=12)
plt.ylabel("Monthly Charges", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x="MonthlyCharges",
    y="TotalCharges",
    data=data,
    hue="Churn",        # Optional: see churn effect
    palette="viridis",
    alpha=0.6
)
plt.title("Monthly Charges vs Total Charges", fontsize=16)
plt.xlabel("Monthly Charges", fontsize=12)
plt.ylabel("Total Charges", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(
    x="MonthlyCharges",
    y="TotalCharges",
    data=data,
    scatter_kws={'alpha':0.3}, 
    line_kws={'color':'red'}
)
plt.title("Regression: Monthly Charges vs Total Charges", fontsize=16)
plt.show()
```

```python
corr_val = data[['MonthlyCharges','TotalCharges']].corr().iloc[0,1]
print("Correlation between MonthlyCharges and TotalCharges:", round(corr_val, 3))
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Scatter Plot: tenure, MonthlyCharges, TotalCharges
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    data['tenure'], 
    data['MonthlyCharges'], 
    data['TotalCharges'], 
    c=pd.Categorical(data['Churn']).codes,  # Color by churn (Yes=1, No=0)
    cmap='viridis',
    alpha=0.6
)

ax.set_xlabel('Tenure')
ax.set_ylabel('Monthly Charges')
ax.set_zlabel('Total Charges')
ax.set_title('3D Plot: Tenure vs MonthlyCharges vs TotalCharges')

plt.colorbar(scatter, ax=ax, label='Churn (0=No, 1=Yes)')
plt.show()
```
