```python
ratings=f_d.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observation
When Rating is between 4.5 to 4.9---> Excellent
When Rating are between 4.0 to 3.4--->very good
when Rating is between 3.5 to 3.9----> good
when Rating is between 3.0 to 3.4----> average
when Rating is between 2.5 to 2.9----> average
when Rating is between 2.0 to 2.4----> Poor


```python
ratings.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

# Set the figure size
matplotlib.rcParams['figure.figsize'] = (12, 6)

# Create colorful barplot
sns.barplot(
    x="Aggregate rating", 
    y="Rating Count", 
    data=ratings, 
    palette="viridis"   # Try: "husl", "pastel", "deep", "coolwarm", etc.
)

plt.title("Rating Count per Aggregate Rating", fontsize=16)
plt.xlabel("Aggregate Rating", fontsize=12)
plt.ylabel("Rating Count", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:

Not Rated count is very high
Maximum number of rating are between 2.5 to 3.4

```python
# ðŸ“Œ Assignment: Find the Top 10 Cuisines

# Drop missing values from 'Cuisines' column to avoid NaN issues
top_cuisines = f_d['Cuisines'].dropna().value_counts().head(10)

# Visualize
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title('Top 10 Cuisines', fontsize=16)
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisines')
plt.tight_layout()
plt.show()
```
